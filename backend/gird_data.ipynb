# Test grid size

```python
# 1. Create a row template
def make_row():
    return [{"value": None, "is_given": False}, 
            {"value": None, "is_given": False}, 
            {"value": None, "is_given": False}]

# Stack rows into a grid
grid = [make_row(), make_row(), make_row()]

# Add clues
grid[0][0]["value"] = 2
grid[0][0]["is_given"] = True

grid[1][2]["value"] = 1
grid[1][2]["is_given"] = True
```

```python
for row in grid:
    print(row)
```

# Grid Data rules

```python
def check_neighbors(r, c):
    val = grid[r][c]["value"]
    if val is None:
        return "Cell is empty"
        
    #Check right cell
    if c + 1 < 3: # don't fall off the edge
        right_neighbor = grid[r][c+1]["value"]
        if right_neighbor == val:
            return "Same number cells must not touch"
    
    return "YAASSS ZADDY"
```
